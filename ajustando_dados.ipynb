### CD-PIB

necessario unificar colunas e pivotar em apenas 1 tabela

```python
CD_pib = arq_xls['CD-PIB']

CD_pib.head()
```

```python
CD_pib_melted = CD_pib.melt(var_name='Var', value_name='val')
```

```python
dict_replace_pib = {
    'Ano.1' : 'Ano',
    'Ano.2' : 'Ano',
    'Ano.3' : 'Ano',
    'PIB.1' : 'PIB',
    'PIB.2' : 'PIB',
    'PIB.3' : 'PIB'
}

CD_pib_melted.replace(dict_replace_pib, inplace=True)
```

```python
# indexando os registros
CD_pib_melted['index'] = CD_pib_melted.groupby('Var').cumcount()

# pivotando os dados
pivot_df = CD_pib_melted.pivot(index='index', columns='Var', values='val')

# resetando o index da tabela
pivot_df = pivot_df.reset_index()

# selecionando apenas as colunas que serao utilizadas
pivot_df = pivot_df[['Ano', 'PIB']]

# trocando os espacos vazios por nan
pivot_df.replace(' ', np.nan, inplace=True)

# removendo campos com ano vazio
pivot_df.dropna(subset=['Ano'], inplace=True)

# transformando o tipo da variavel
pivot_df['Ano'] = pivot_df['Ano'].astype('int')

# aplicando na base final
CD_pib = pivot_df

CD_pib.head()
```

### CD-Mercado

Dados com a mesma divisao em colunas da base anterior, necessario unificar colunas e pivotar em apenas 1 tabela

```python
# puxando df
CD_mercado = arq_xls['CD-Mercado']

# separando as colunas do dataset
_subset1 = CD_mercado.iloc[:, :4]
_subset2 = CD_mercado.iloc[:, 4:8]
_subset3 = CD_mercado.iloc[:, 8:12]
```

```python
# nome desejado das colunas
_nomes_col_cd_mercado = ['mes', 'dia', 'telebras', 'indice']

# listando os dataframes
lista_subsets = [_subset1, _subset2, _subset3]

# iterando o nome das colunas desejados para cada dataframe
for s in lista_subsets:
    s.columns = _nomes_col_cd_mercado

# checando por linhas inteiras vazias e removendo
for s in lista_subsets:
    s.dropna(how='all', inplace=True)

# concatenando os datasets
CD_mercado = pd.concat(lista_subsets)

# preenchendo a coluna de meses
CD_mercado['mes'].fillna(method='ffill', inplace=True)

CD_mercado.head()
```

```python
# listando todos os datasets do livro tratados e exportando csv

_datasets_estat_bas = [CD_tb21, CD_Brasil, CD_Municipios, CD_notas, CD_poluicao, CD_temperaturas, CD_salarios, 
                       CD_veiculos, CD_pib, CD_mercado, CD_placa_bac, CD_manchas_solares, CD_precipitacao, CD_co2]

# iterando datasets
for i in _datasets_estat_bas:

    # pegando o nome da variavel que origina o dataframe
    nome_tbl = [name for name, data in globals().items() if data is i][0]

    # nomeando arquivo e exportando como csv
    nome_arquivo = f'{nome_tbl}.csv'
    i.to_csv(f'base_dados_EB/{nome_arquivo}', index = False)

    print(f'dataset {nome_tbl} exportado')
```
