### Exercício 2

```python
# importando os dados do exercicio
df_tb21 = pd.read_csv('base_dados_EB/tabela2_1.csv')
```

```python
df_tb21.head()
```

```python
# Distribuicao de frequencia - estado civil
df_tb21_agrupado_a = df_tb21.groupby('Estado Civil').size().reset_index(name = 'Frequencia')

# criando a coluna de porcentagem
df_tb21_agrupado_a['Porcentagem'] = ((df_tb21_agrupado_a['Frequencia'] / df_tb21_agrupado_a['Frequencia'].sum()) * 100).round(2)

df_tb21_agrupado_a
```

```python
# Distribuicao de frequencia - Regiao procedencia
df_tb21_agrupado_b = df_tb21.groupby('Região de Procedência').size().reset_index(name = 'Frequencia')

# criando a coluna de porcentagem
df_tb21_agrupado_b['Porcentagem'] = ((df_tb21_agrupado_b['Frequencia'] / df_tb21_agrupado_b['Frequencia'].sum()) * 100).round(2)

df_tb21_agrupado_b
```

```python
# Distribuicao de frequencia - Regiao procedencia
df_tb21_agrupado_c = df_tb21[df_tb21['Estado Civil'] == 'casado'].groupby('N de Filhos').size().reset_index(name = 'Frequencia')

# criando a coluna de porcentagem
df_tb21_agrupado_c['Porcentagem'] = ((df_tb21_agrupado_c['Frequencia'] / df_tb21_agrupado_c['Frequencia'].sum()) * 100).round(2)

df_tb21_agrupado_c
```

```python
# Distribuicao de frequencia - Regiao procedencia
df_tb21_agrupado_d = df_tb21.groupby('Anos').size().reset_index(name = 'Frequencia')

# criando a coluna de porcentagem
df_tb21_agrupado_d['Porcentagem'] = ((df_tb21_agrupado_d['Frequencia'] / df_tb21_agrupado_d['Frequencia'].sum()) * 100).round(2)

df_tb21_agrupado_d
```

### Exercício 3

Usando o conjunto de dados CD-Brasil, construa a distribuição de frequências para as variáveis população e densidade populacional

```python
df_cd_brasil = pd.read_csv('base_dados_EB\CD_Brasil.csv')

df_cd_brasil.head()
```

```python
# populacao em mil habitantes
df_cd_brasil['pop1000'] = df_cd_brasil['populacao']/1000

# agrupar os dados de populacao

df_cd_brasil['fx_pop'] = np.where(df_cd_brasil['pop1000'] < 1000, 'Menor que 1000',
                         np.where(df_cd_brasil['pop1000'] < 5000, '1000 - 5000',
                         np.where(df_cd_brasil['pop1000'] < 10000, '5000 - 10000',
                         np.where(df_cd_brasil['pop1000'] < 20000, '10000 - 20000',
                         np.where(df_cd_brasil['pop1000'] > 20000, 'Acima de 20000', 'a')))))

# contando as registros das faixas
df_cd_brasil1 = df_cd_brasil.groupby('fx_pop').size().reset_index(name = 'freq')

# coluna de porcentagem
df_cd_brasil1['Porcentagem'] = ((df_cd_brasil1['freq'] / df_cd_brasil1['freq'].sum()) * 100).round(2)

# listando a ordem desejada da coluna fx_pop
ordem_fx_pop = ['Menor que 1000', '1000 - 5000', '5000 - 10000', '10000 - 20000', 'Acima de 20000']

# transformando a variavel em categorica e aplicando a ordem desejada na lista anterior
df_cd_brasil1['fx_pop'] = pd.Categorical(df_cd_brasil1['fx_pop'], categories=ordem_fx_pop, ordered=True)

# ordenando a tabela pela faixa desejada
df_cd_brasil1 = df_cd_brasil1.sort_values('fx_pop')

df_cd_brasil1
```

```python
df_cd_brasil.head()
```

```python
# calculando a densidade populacional
# total de habitantes / pela area (superficie)

df_cd_brasil['dens_pop'] = df_cd_brasil['populacao']/df_cd_brasil['superficie']

# dividindo a densidade populacional em faixas

df_cd_brasil['fx_dens_pop'] = np.where(df_cd_brasil['dens_pop'] < 10, 'Menor que 10',
                              np.where(df_cd_brasil['dens_pop'] < 30, '10 - 30',
                              np.where(df_cd_brasil['dens_pop'] < 50, '30 - 50',
                              np.where(df_cd_brasil['dens_pop'] < 100, '50 - 100', 'Acima de 100'))))

# contando as registros das faixas
df_cd_brasil2 = df_cd_brasil.groupby('fx_dens_pop').size().reset_index(name = 'freq')

# coluna de porcentagem
df_cd_brasil2['Porcentagem'] = ((df_cd_brasil2['freq'] / df_cd_brasil2['freq'].sum()) * 100).round(2)

# listando a ordem desejada da coluna fx_pop
ordem_fx_dens_pop = ['Menor que 10', '10 - 30', '30 - 50', '50 - 100', 'Acima de 100']

# transformando a variavel em categorica e aplicando a ordem desejada na lista anterior
df_cd_brasil2['fx_dens_pop'] = pd.Categorical(df_cd_brasil2['fx_dens_pop'], categories=ordem_fx_dens_pop, ordered=True)

# ordenando a tabela pela faixa desejada
df_cd_brasil2 = df_cd_brasil2.sort_values('fx_dens_pop')

df_cd_brasil2
```
