### CD-Temperaturas

tabela multi indice e com repeticoes de colunas

```python
CD_temperaturas = arq_xls['CD-Temperaturas']

CD_temperaturas.head()
```

```python
# selecionando as primeiras 3 colunas
CD_temperaturas1 = CD_temperaturas.iloc[:, :3]
CD_temperaturas1.replace(' ', np.NaN, inplace= True)
CD_temperaturas1.fillna(method='ffill', inplace=True)

# ultimas 3 colunas
CD_temperaturas2 = CD_temperaturas.iloc[:, -3:]
CD_temperaturas2.replace(' ', np.NaN, inplace=True)
CD_temperaturas2.fillna(method='ffill', inplace=True)

# unificando os nomes
CD_temperaturas2.columns = ['Ano', 'Cananéia', 'Ubatuba']

# empilhando em uma unica tabela
CD_temperaturas = pd.concat([CD_temperaturas1, CD_temperaturas2], ignore_index=True)
```

### CD-PIB

necessario unificar colunas e pivotar em apenas 1 tabela

```python
CD_pib = arq_xls['CD-PIB']

CD_pib.head()
```

```python
CD_pib_melted = CD_pib.melt(var_name='Var', value_name='val')
```

```python
dict_replace_pib = {
    'Ano.1' : 'Ano',
    'Ano.2' : 'Ano',
    'Ano.3' : 'Ano',
    'PIB.1' : 'PIB',
    'PIB.2' : 'PIB',
    'PIB.3' : 'PIB'
}

CD_pib_melted.replace(dict_replace_pib, inplace=True)
```

```python
# indexando os registros
CD_pib_melted['index'] = CD_pib_melted.groupby('Var').cumcount()

# pivotando os dados
pivot_df = CD_pib_melted.pivot(index='index', columns='Var', values='val')

# resetando o index da tabela
pivot_df = pivot_df.reset_index()

# selecionando apenas as colunas que serao utilizadas
pivot_df = pivot_df[['Ano', 'PIB']]

# trocando os espacos vazios por nan
pivot_df.replace(' ', np.nan, inplace=True)

# removendo campos com ano vazio
pivot_df.dropna(subset=['Ano'], inplace=True)

# transformando o tipo da variavel
pivot_df['Ano'] = pivot_df['Ano'].astype('int')

# aplicando na base final
CD_pib = pivot_df

CD_pib.head()
```
