### CD-Notas

Dataset contem notas de 100 alunos do curso de econimia da FEA-USP.

deve-se transformar todas as notas em apenas 1 coluna

```python
CD_notas = arq_xls['CD-Notas']

CD_notas.head()
```

```python
# obtendo os valores do header
header_cd_notas = CD_notas.columns.tolist()

# renomeando as colunas (indices)
CD_notas.columns = range(10)

# adicionando as notas no dataset
CD_notas.loc[len(CD_notas)] = header_cd_notas

#transformando as notas em uma unica colunas
CD_notas = CD_notas.melt(value_vars=CD_notas.columns).drop(columns='variable')
```

```python
CD_notas.head()
```

### CD-Temperaturas

tabela multi indice e com repeticoes de colunas

```python
CD_temperaturas = arq_xls['CD-Temperaturas']

CD_temperaturas.head()
```

```python
# selecionando as primeiras 3 colunas
CD_temperaturas1 = CD_temperaturas.iloc[:, :3]
CD_temperaturas1.replace(' ', np.NaN, inplace= True)
CD_temperaturas1.fillna(method='ffill', inplace=True)

# ultimas 3 colunas
CD_temperaturas2 = CD_temperaturas.iloc[:, -3:]
CD_temperaturas2.replace(' ', np.NaN, inplace=True)
CD_temperaturas2.fillna(method='ffill', inplace=True)

# unificando os nomes
CD_temperaturas2.columns = ['Ano', 'Cananéia', 'Ubatuba']

# empilhando em uma unica tabela
CD_temperaturas = pd.concat([CD_temperaturas1, CD_temperaturas2], ignore_index=True)
```
